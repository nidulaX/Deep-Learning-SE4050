```python

# Create x and y values in range [-5, 5]
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)

# Equation Z = X^2 + Y^2
Z = X**2 + Y**2

# Plotting
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap=matplotlib.cm.coolwarm, linewidth=0.1)

# Labels
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z = XÂ² + YÂ²")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the online store customer data dataset
url = "/content/drive/MyDrive/Colab Notebooks/online_store_customer_data.csv"
df = pd.read_csv(url)

# Select numerical features for correlation analysis
# Based on the dataframe inspection, 'Age', 'Referal', and 'Amount_spent' are numerical
features = ['Age', 'Referal', 'Amount_spent']
data = df[features]

# Calculate Pearson correlation
pearson_corr = data.corr(method='pearson')

# Calculate Spearman correlation
spearman_corr = data.corr(method='spearman')

# Plot heatmaps
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Pearson Correlation")

plt.subplot(1, 2, 2)
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Spearman Correlation")

plt.tight_layout()
plt.show()
```

```python
#Create a new dataframe by filtering all the columns [i.e., features] except N_total, N_bulk, N_surface and R_avg columns.
new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
```

```python
# Display the first 20 samples of this dataframe.
print("First 20 samples of the dataframe:")
display(new_df.head(20))
```

```python
# Step 4: Calculate the mean, standard deviation and quartile values for each of the above 4 features.
stats_df = pd.DataFrame({
    'Mean': new_df.mean(),
    'Standard Deviation': new_df.std(),
    '25% Quantile': new_df.quantile(0.25),
    '50% Quantile (Median)': new_df.quantile(0.50),
    '75% Quantile': new_df.quantile(0.75)
})
print("\nStatistical Summary:")
display(stats_df)
```

```python
# Step 5: Plot histogram for each feature in a 1x4 layout
plt.figure(figsize=(20, 4))
for i, col in enumerate(new_df.columns):
    plt.subplot(1, 4, i+1)
    plt.hist(new_df[col], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {col}')
plt.tight_layout()
plt.show()
```

```python
# Step 6: Seaborn pairplot (scatter plots + histograms)
sns.pairplot(new_df)
plt.show()
```

```python
# Step 7: Custom PairGrid
g = sns.PairGrid(new_df)

# Upper half: bivariate histogram
g.map_upper(sns.histplot)

# Diagonal: histogram + KDE
g.map_diag(sns.histplot, kde=True)

# Lower half: bivariate KDE
g.map_lower(sns.kdeplot, fill=True)

plt.show()
```
